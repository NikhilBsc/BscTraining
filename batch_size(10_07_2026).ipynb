```python
#Gradient descent
import numpy as np
X=np.array([[1],[2]])
y=np.array([[2],[4]])
weight = 0.5
bias = 0.1
learning_rate = 0.1
for epochs in range(10):
  y_pred = X* weight + bias
  loss=np.mean((y - y_pred)**2)
  dw =(-2/len(X)) * np.sum(X*(y-y_pred))
  db =(-2/len(X)) * np.sum(y-y_pred)
  weight = weight - learning_rate * dw
  bias = bias - learning_rate * db
  print(f"Epoch : {epochs + 1}")
  print("loss,round(loss,4)")

```

#Stochastic GD

```python
X= np.array([1,2,3,4])
y=np.array([2,4,6,8])

weight =0.5
bias = 0.1
learning_rate = 0.1

for X, y in zip(X,y):
  y_pred = weight * X + bias
  error = y - y_pred
  dw = -2 * X * (y - y_pred)
  db = -2 * error

  weight = weight - learning_rate*dw

  bias = bias -learning_rate * db

  print(f"input = {X}, Actual ={y}")
  print(f"prediction = {round(y_pred,2)}")
  print(f"updated weight = {round(weight,2)}")
  print(f"updated bias = {round(bias,2)}")
print(f"final weight = {round(weight,2)}")
```

```python
#Adam

import numpy as np

X = np.array([[1],[2]])
y = np.array([[2],[4]])

weight = np.array([[0.5]])
bias = np.array([[0.1]])
```

```python
learning_rate = 0.1

for epoch in range(10):
  y_pred =np.dot(X,weight)+bias

  loss = np.mean((y-y_pred)**2)

  dw = (-2/len(X))* np.dot(X.T,(y-y_pred))
  db = (-2/len(X)*np.sum(y-y_pred))

  weight =weight - learning_rate
  bias = bias -learning_rate
  print(f"epoch {epoch +1}")
```

```python
# Binary step function
import numpy as np
import matplotlib.pyplot as plt

def binary_step(x):
  return np.where(x>=0 , 1,0)
x= np.linspace(-5,5,100)
y=binary_step(x)
print("input :",x[:10])
print("output :",y[:10])

plt.plot(x,y,linewidth =3)
plt.title("binary step activation")
plt.grid(True)
plt.show()
```

```python
#linear activation function
import numpy as np
import matplotlib.pyplot as plt

def linear(x):
  return (x)

x=np.linspace(-5,5,100)
y=linear(x)

print(y[:10])

plt.plot(x,y,linewidth =3)
plt.title("linear activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid(True)
plt.show()
```

```python
#sigmoid function
def sigmoid(x):
  return 1/(1+np.exp(-x))

x= np.linspace(-10,10,100)
y=sigmoid(x)

print(y[:10])

plt.plot(x,y , linewidth=3)
plt.title("sigmoid activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid(True)
plt.show()
```

```python
#Tanh
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
  return np.tanh(x)
x = np.linspace(-5 ,5,100)
y=tanh(x)

print(y[:10])

plt.plot(x,y,linewidth = 3)
plt.title("Tanh Activation Function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid(True)
plt.show()
```

```python

#relu activn function
def relu(x):
  return np.maximum(0 , x)

x= np.linspace(-5,5,100)
y=relu(x)

print(y[:10])

plt.plot(x,y,linewidth = 3)
plt.title("relu Activation Function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid(True)
plt.show()
```

```python
#softmax activation

def softmax(x):
  exp=np.exp(x)
  return exp /np.sum(exp)

x=np.array([2.0,1.0,0.1])
y=softmax(x)
print("scores :",x)
print("prob ",y)

plt.bar(["class A","class b","class c"],y)
plt.title("softmax otput")
plt.show()
```
