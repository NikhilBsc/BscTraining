#23
A 10 *10 image is passed through a Conv2D layer with a 3 *3 filter, stride 1, and no padding. Using the formula (N−F+2P)/S+1, the output size becomes (10−3+0)/1+1=8. Therefore, the Conv2D layer produces an 8 * 8 feature map. A 2 * 2 MaxPooling layer then reduces the dimensions by half, resulting in a final output shape of 4 * 4.

```python

#24
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    MaxPooling2D,
    Flatten,
    Dense,
    Dropout,
    BatchNormalization
)

model = Sequential([

    Conv2D(
        32,
        (3,3),
        activation='relu',
        input_shape=(224,224,1)
    ),

    BatchNormalization(),

    MaxPooling2D((2,2)),

    Conv2D(
        64,
        (3,3),
        activation='relu'
    ),

    BatchNormalization(),MaxPooling2D((2,2)),Flatten(),
    Dense(
        128,
        activation='relu'
    ),
    Dropout(0.5),
    Dense(2,activation='softmax')
])

model.summary()
```
