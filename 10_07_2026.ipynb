```python
#q1
The neural network consists of three layers:

Input Layer: Contains 3 input neurons (x₁, x₂, x₃) that receive the input data.
Hidden Layer: Contains 4 neurons. Each neuron receives information from all three input neurons and  processes it.
Output Layer: Contains 1 neuron that produces the final prediction or output.

So, the architecture  can be represented as:

3 Input Neurons -> 4 Hidden Neurons -> 1 Output Neuron
```

```python
#q2
Between the input layer and hidden layer:

3 inputs * 4 hidden neurons = 12 weights

Between the hidden layer and output layer:

4 hidden neurons * 1 output neuron = 4 weights

Total Weights = 12 + 4 = 16

For biases:

Hidden layer = 4 biases (one for each hidden neuron)
Output layer = 1 bias

Total Biases = 5
```

```python
#q5
A weight determines how much influence an input has on the neurons output.inputs with larger weights contribute more to the final output

A bias is an additional value added to the weighted sum. It helps the neuron adjust its output even when all input values are zero.

In simple terms, weights control the importance of inputs, while bias helps shift the output to improve learning.
```

```python
#q6

If all weights are initialized to zero, every neuron in the same layer will produce the same output and receive the same updates during training. As a result, all neurons learn the same features and the network cannot learn effectively.

This problem is known as the symmetry problem, which is why neural networks usually initialize weights with small random values.
```

```python
#q7 (components of neuron)
Inputs - The data values provided to the neuron.
Weights - Values that determine the importance of each input.
Bias - An additional parameter added to the weighted sum.
Activation Function - Converts the weighted sum into the final output of the neuron.
```

```python
#q9
import pandas as pd
data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7, 8, 2, 4, 6, 8],
    'Exam_Score': [35, 40, 45, 50, 60, 65, 70, 80, 38, 55, 68, 85]
}
df = pd.DataFrame(data)
print(df)
```

#q10
The dataset contains 12 samples and the batch size is 3.

Number of batches = 12 / 3 = 4

So, 4 batches will be created.

```python
batch_size = 3

for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i+batch_size]
    print(f"\nBatch {(i // batch_size) + 1}")
    print(batch)
```

```python
#12
When the batch size is equal to the entire dataset, it is called Batch Gradient Descent.

The model processes all training samples before updating the weights
```

```python
#q13
When the batch size is 1, it is called Stochastic Gradient Descent (SGD).

The model updates its weights after processing each individual sample.
```

Example: Cat, Dog, and Bird classification.

```python
#46
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = 2
print("Sigmoid(2) =", sigmoid(x))
```

```python
#47
import numpy as np
def relu(x):
    return np.maximum(0, x)

values = [-5, -2, 0, 3, 7]
print("Input:", values)
print("ReLU Output:", relu(values))
```
