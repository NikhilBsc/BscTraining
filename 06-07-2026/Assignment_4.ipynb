```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = "https://books.toscrape.com/"
response = requests.get(url)
print("Status Code :", response.status_code)
soup = BeautifulSoup(response.text, "html.parser")
books = soup.find_all("article", class_="product_pod")
print("Total Books :", len(books))
```

```python
book_name=[]
price=[]
rating=[]
availability=[]
for book in books:
  name=book.h3.a["title"]
  cost=book.find("p",class_="price_color").txt
  stars=book.find("p")["class"][1]
  stock=book.find("p",class_="instock availability").text.strip()
  book_name.append(name)
  price.append(cost)
  rating.append(stars)
  availability.append(stock)
  print(book_name)
  print(price)
  print(rating)
```

#Remote job sarch , WebScrapping example

```python
import requests
from bs4 import BeautifulSoup
job_data=[]

url = "https://realpython.github.io/fake-jobs/"

response1 = requests.get(url)
print("status code",requests.status_codes)

soup1 = BeautifulSoup(response1.text, "html.parser")

print("Status:", response1.status_code)
jobs = soup1.find_all("div", class_="card-content")
print("Total Jobs:", len(jobs))


```

```python
for job in jobs[:5]:
    title = job.find("h2", class_="title is-5")
    print(title.text.strip())
```

```python
for job in jobs[:5]:
    company = job.find("h3", class_="subtitle is-6 company")
    print(company.text.strip())
```

```python
for job in jobs[:5]:
    location = job.find("p", class_="location")
    print(location.text.strip())
```

```python
for job in jobs[:5]:
    date = job.find("time")
    print(date.text.strip())
```

```python
import pandas as pd
job_data = []
df = pd.DataFrame(job_data,columns=["Job Title","Company","Location","Date"]
)
print(df.head())
```

```python
import requests
import pandas as pd
# API endpoint
url = "https://jsonplaceholder.typicode.com/users"
# Sending GET Request
response = requests.get(url)
# Check response status
if response.status_code == 200:
    # Convert JSON to Python way
    data = response.json()
    # Create DataFrame
    df = pd.DataFrame(data)
    print("data retrieved succesfully")
    print(df.head())
    # Saving as CSV
    df.to_csv("users.csv", index=False)
    # Saving  as Excel
    df.to_excel("users.xlsx", index=False)
    print("files saved succesfully")
else:
    print("Error:", response.status_code)
```
