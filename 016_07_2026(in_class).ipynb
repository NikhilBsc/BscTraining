```python
sentences = [ "i love NLP","I love python","NLP is fun"]

vocabulary = set()

for sentence in sentences:
  words = sentence.lower().split()

  for word in words:
    vocabulary.add(word)

print(vocabulary)
```

```python
sentences = [ "i love NLP","I love python","NLP is fun"]

vocabulary = sorted(set(word
 for sentence in sentences
 for word in sentence.lower().split()))

word_to_index = {}

for index , word in enumerate(vocabulary):
  word_to_index[word] = index
print(word_to_index)
```

```python
#creating one - hot vectors
vocbulary =["i","love","nlp","pyhton"]

word_to_index = {"i":0,"love":1,"nlp":2,"python":3}

word = "love"
vectoe =[0] * len(vocbulary)
vectoe[word_to_index[word]] = 1
print(vectoe)
```

```python
#manual bag of words
sentences = ["I love python","i love nlp"]
vocabulary = sorted(set(word
 for sentence in sentences
 for word in sentence.lower().split()))

for sentence in sentences :
  vector =[0]*len(vocabulary)
  words = sentence.lower().split()

  for word in words:
    index = vocabulary.index(word)
    vector[index] += 1
  print(sentence)
  print(vector)
```

```python
#CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

documents =["i love python","i love nlp","python is amazing"]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
#N -grams with CountVectorizer
documents =["Machine learning is  fun","i love machine learning"]

#unigram only
CountVectorizer(ngram_range=(1,1))

#bigram only
CountVectorizer(ngram_range=(2,2))

#unigram and bigram
CountVectorizer(ngram_range=(1,2))
#
```

```python
#TF - IDF using Scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

docs =["i love pyhton","i love nlp","python is amazing"]

vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(docs)

print(vectorizer.get_feature_names_out())
print(x.toarray())
```

#tokenizing text


```python
from transformers import AutoTokenizer
tokenizer =AutoTokenizer.from_pretrained("gpt2")
text = "Ar tificial intelligence is amzing"
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
#convert tokens to IDS
token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(token_ids)
```

```python
#encoding and decoding
encoded = tokenizer(text)
print(encoded)
```

```python
decoded = tokenizer.decode(encoded["input_ids"])
print(decoded)
```
