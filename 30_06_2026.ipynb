```python
#Spam detection function
def detect_spam(message):
    spam_words = ["win", "prize", "free", "offer", "money","lottery","congratulations","winner"]
    message = message.lower()
    for word in spam_words:
        if word in message:
            return "spam"
    return "not spam"
message=input("enter the message")
detect_spam(message)
```

#constructor
They are special methods that gets called when an object is being cretaed . constructors are used to give values to an object when it is created. this saves time because we do not have to assign values separately after creating the object.

for example, when creating a student object, we can directly pass the students name and course using the constructor instead of assigning them later.

```python
#Encapsulation
class BankAccount:
    def __init__(self):
        self.__balance = 0
    def deposit(self, amount):
        self.__balance += amount
    def withdraw(self, amount):
        self.__balance -= amount
    def check_balance(self):
        print("balance:", self.__balance)
acc = BankAccount()
acc.deposit(5000)
acc.withdraw(1000)
acc.check_balance()
```

```python
#30
import logging
logging.basicConfig(
    filename="ai_system.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)
class SpamDetector:
    def __init__(self, model_name, version):
        self.model_name = model_name
        self.version = version
    def predict(self, message):
        if message == "":
            raise ValueError("message cannot be empty")
        spam_words = ["free", "offer", "win","lottery","win"]
        for word in spam_words:
            if word in message.lower():
                return "Spam"
        return "Not Spam"

detector = SpamDetector("spam detector", "v1")
try:
    message = input("enter message: ")

    result = detector.predict(message)

    file = open("prediction.txt", "a")
    file.write(result + "\n")
    file.close()
    logging.info(f"Prediction : {result}  Confidence : 94%")

    print("result:", result)
except ValueError as e:
    print(e)
```

#35 I would use file handling. Whenever a student object is created, I would save the student's information in a text file. Even after the program closes, the data remains in the file and can be read again later.

```python
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course
    def save(self):
        file = open("students1.txt", "a")

        file.write(self.name + " - " + self.course + "\n")

        file.close()
name = input("enter name: ")
course = input("enter course: ")
s1 = Student(name, course)
s1.save()
print("student saved")
```
