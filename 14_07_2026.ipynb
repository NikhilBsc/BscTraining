#1
Natural Language Processing (NLP) is a branch of Artificial Intelligence that helps computers understand, interpret, and generate human language. It acts as a bridge between humans and computers by converting human language into a form that computers can process.

#2
Computers understand numbers and code, not human language. NLP helps convert text or speech into a format that computers can analyze.

Example: Google Translate uses NLP to translate text from one language to another.

#q3
voice Assistants – Applications like Siri and Google Assistant understand voice commands and respond to users.
Text Autocomplete – Suggests words or sentences while typing in emails, messages, or search engines.
Grammar and Spell Checkers – Tools like Grammarly detect spelling and grammar mistakes and suggest corrections.
Text Summarization – Automatically creates a short summary of long articles, documents, or reports.

#q4
A rule-based NLP system follows predefined rules to make decisions.

Example: In spam detection, if a message contains the word "Congratulations", it may be marked as spam based on a fixed rule.

#5
Text Collection – Gather text data.
Text Cleaning – Remove unwanted characters and symbols.
Tokenization – Split text into words or tokens.
Lowercasing – Convert all text to lowercase.
Stopword Removal – Remove common words like "the", "is".
Stemming/Lemmatization – Convert words to their root form.
Feature Extraction & Modeling – Convert text into numbers and train a model.

#q6
okenization is the process of splitting text into smaller units called tokens (usually words).

Example:
"I love NLP" ---> ["I", "love", "NLP"]

It is necessary because machine learning models cannot directly understand raw text.

#q7
stopwords are common words that usually do not add much meaning to a sentence.

Examples: "the", "is", "and"

They are often removed to reduce unnecessary data and improve processing speed.

#q8
Lowercasing ensures that words with different cases are treated as the same word.

Example:
"Apple", "APPLE", and "apple" will all become "apple".

This avoids duplicate representations of the same word.
